# **Crawling**

```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import time
query ="("+Kata_Kunci+")"+ " exclude:nativeretweets lang:id until:"+Akhir_Tanggal+"since"+Start_Tanggal
tweets = []
limit = Jumlah

start = time.time()
try :
    for tweet in sntwitter.TwitterSearchScraper(query=query).get_items ():
        if len (tweets) ==limit :
            break
        else :
            tweets.append ([tweet.date,tweet.user.username,tweet.content,tweet.likeCount,tweet.retweetCount])
    Dataset = pd.DataFrame (tweets, columns = ['Date','Username','Tweet','Likes','Retweets'])

except Exception as e :
    print (e)

end = time.time()
print(f"Duration : {end - start} Seconds")
```

```python
Dataset = Dataset.sort_values(by=['Retweets'],ascending = False )
Dataset['Date'] = Dataset['Date'].dt.tz_localize(None)
Dataset
```

```python
Dataset.to_excel ('ReportCrawlingPertanian.xlsx',index =False)
```

# **Most Word**

```python
Data = pd.read_excel ('ReportCrawlingPertanian.xlsx')
Data
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
import string
import emoji
import re
from collections import Counter
import nltk
nltk.download('punkt')
import requests


def preprocessing(tweets):
    tweets = tweets.encode('ascii','ignore').decode('utf-8')
    tweets = re.sub(r'[^\x00-\x7f]',r'',tweets)
    tweets = re.sub(r'[_.(){}[]]+','',tweets)
    tweets = re.sub(r'@[A-Za-z0-9]+','',tweets)
    tweets = re.sub(r'#[A-Za-z0-9]+','',tweets)
    tweets = re.sub(r'_[A-Za-z0-9]+','',tweets)
    tweets = re.sub(r'-[A-Za-z0-9]+','',tweets)
    tweets = re.sub(r':[A-Za-z0-9]+','',tweets)
    tweets = re.sub(r'https?:\/\/\S+','',tweets)
    tweets = re.sub(r'\d+','',tweets)
    tweets = re.sub(r'@+','',tweets)
    tweets = re.sub(r':+','',tweets)
    tweets = re.sub(r',+','',tweets)
    tweets = re.sub(r'_+','',tweets)
    tweets = tweets.lower()

    allemot = [str for str in tweets]
    listemot = [x for x in allemot if x in emoji.distinct_emoji_list(tweets)]
    tweets = ' '.join([str for str in tweets.split()if not any (y in str for str in listemot)])
    return tweets
Data['Tweet'] = Data['Tweet'].apply(preprocessing)
Data
```

```python
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

Data['Tweet']  = Data['Tweet'].apply(remove_punctuations)
Data
```

```python
url = "https://raw.githubusercontent.com/mychaelgo/indonesia-wordlist/master/indonesian-wordlist.lst"
response = requests.get(url)

if response.status_code == 200:
    word_list = response.content.decode("utf-8").split("\n")
    print(word_list[:10])
else:
    print("Failed to retrieve the wordlist.")
```

```python
factory = StopWordRemoverFactory().get_stop_words()
more_stopword = word_list
data = factory + more_stopword
dictionary = ArrayDictionary(data)
stp = StopWordRemover(dictionary)

text_stopwords = []
for index, row in Data.iterrows():
    text_stopwords.append(stp.remove(row['Tweet']))

Data['Tweet'] = text_stopwords
Data
```

```python
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

text_stopwords = []
for index, row in Data.iterrows():
    text_stopwords.append(stopword.remove(row['Tweet']))

Data['Tweet'] = text_stopwords
Data
```

```python
tweet_data = Data["Tweet"]
all_tweets = ' '.join(tweet_data)
words = all_tweets.split()
word_counts = Counter(words)

most_common_words = pd.DataFrame(word_counts.most_common(Kata_kunci), columns=["Word", "Frequency"])
most_common_words
```
